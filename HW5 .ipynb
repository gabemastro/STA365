## Question 2 

### **Problem Overview**
We run Bayesian inference using PyMC for three different prior choices on $\theta$ and $\tau$ , then analyze the outputs using:
- **Trace plots**
- **Posterior distributions**
- **Autocorrelation plots**
- **Energy plots for rejection rates**
- **Forest plots**
- **R-hat (Gelman-Rubin) diagnostic summary statistics**

```python
import numpy as np
import pymc as pm
import arviz as az
import matplotlib.pyplot as plt

# Generate synthetic data
np.random.seed(42)
n = 50  # Sample size
m = 2   # Mean of the normal distribution
s = 3   # Standard deviation of the normal distribution
x = np.random.normal(loc=m, scale=s, size=n)

# Define three Bayesian models for different priors on theta and tau
def bayesian_inference(case):
    with pm.Model() as model:
        
        # Define priors based on the case
        if case == 1:
            theta = pm.Normal("theta", mu=0, sigma=1)  # Normal prior for theta
            tau = pm.Gamma("tau", alpha=2, beta=1)  # Gamma prior for tau
        elif case == 2:
            theta = pm.Laplace("theta", mu=0, b=1)  # Laplace prior for theta
            tau = pm.Exponential("tau", lam=1)  # Exponential prior for tau
        elif case == 3:
            theta = pm.StudentT("theta", nu=5, mu=0, sigma=1)  # Student-T prior for theta
            tau = pm.InverseGamma("tau", alpha=2, beta=1)  # Inverse-Gamma prior for tau
        
        # Likelihood
        y_obs = pm.Normal("y_obs", mu=theta, sigma=1/pm.math.sqrt(tau), observed=x)
        
        # Perform inference using MCMC
        trace = pm.sample(2000, tune=1000, target_accept=0.9, return_inferencedata=True)
    
    return trace

# Run models for all three cases and process each sequentially
for i in range(1, 4):
    print(f"\nProcessing Case {i}...\n")

    trace = bayesian_inference(i)

    # Trace and Posterior plots
    az.plot_trace(trace)
    plt.suptitle(f"Case {i}: Trace Plot", fontsize=14)
    plt.show()
    
    az.plot_posterior(trace)
    plt.suptitle(f"Case {i}: Posterior Distribution", fontsize=14)
    plt.show()
    
    # Compute and plot autocorrelations for theta and tau in one plot
    fig, ax = plt.subplots(figsize=(7, 5))

    # Number of lags
    max_lag = 30

    # Extract theta and tau samples
    theta_samples = trace.posterior["theta"].values.flatten()
    tau_samples = trace.posterior["tau"].values.flatten()

    # Compute autocorrelations
    theta_acf = [np.corrcoef(theta_samples[:-lag], theta_samples[lag:])[0, 1] for lag in range(1, max_lag)]
    tau_acf = [np.corrcoef(tau_samples[:-lag], tau_samples[lag:])[0, 1] for lag in range(1, max_lag)]

    # Plot autocorrelations
    ax.plot(range(1, max_lag), theta_acf, label=" (Theta)")
    ax.plot(range(1, max_lag), tau_acf, label=" (Tau)")

    # Formatting
    ax.axhline(0, color='black', linewidth=0.5, linestyle="--")
    ax.set_xlabel("Lag")
    ax.set_ylabel("Autocorrelation")
    ax.set_title(f"Case {i}: Autocorrelation of  and ")
    ax.legend()

    plt.show()

    # Energy plot for rejection rates
    az.plot_energy(trace)
    plt.suptitle(f"Case {i}: Energy Plot (Rejection Rates)", fontsize=14)
    plt.show()
    
    # Forest plot for diagnostics
    az.plot_forest(trace, var_names=["theta", "tau"], combined=True, hdi_prob=0.95, r_hat=True)
    plt.suptitle(f"Case {i}: Forest Plot (Diagnostics)", fontsize=14)
    plt.show()
    
    # Print summary statistics
    summary = az.summary(trace, round_to=5)
    print(summary)
    
    # Check convergence diagnostics (Gelman-Rubin statistic R-hat)
    r_hat_values = az.rhat(trace)
    print(f"\nGelman-Rubin R-hat values for Case {i}:")
    print(r_hat_values)
    print("-" * 60)
```

## Question 3 

### Markov Algorithm of Slice Sampling
Slice sampling is an **MCMC (Markov Chain Monte Carlo) method** used to sample from a probability distribution when direct sampling is difficult. It introduces an auxiliary variable to define a slice of the probability distribution and then samples alternately from the horizontal and vertical dimensions.

### Steps of Slice Sampling:
1. **Initialize** at some point $x_t$.
2. **Sample a vertical threshold** $y$ uniformly:
   $$ y \sim \text{Uniform}(0, f(x_t)) $$
   This defines the horizontal slice where the next sample will be drawn.
3. **Find an interval** $(L, R)$ containing $x_t$ where:
   $$ f(x) > y, \quad \forall x \in (L, R) $$
   This is typically done using an expansion step.
4. **Sample a new point** $x_{t+1}$ uniformly from the valid interval:
   $$ x_{t+1} \sim \text{Uniform}(L, R) $$
   If the function $f(x_{t+1}) < y$, the interval is shrunk accordingly.
5. **Repeat the process** to generate a Markov chain.

## Slice Sampling in Metropolis-Within-Gibbs
Metropolis-Within-Gibbs sampling is used when full conditional distributions are known only up to a normalizing constant. Instead of using a standard Metropolis-Hastings step, we can use slice sampling as follows:

1. **Define the full conditional distribution** for a parameter $\theta$:
   $$ p(\theta \mid \text{rest}) \propto f(\theta) $$
   but the normalizing constant is unknown.
2. **Introduce an auxiliary variable** $y$:
   $$ y \sim \text{Uniform}(0, f(\theta)) $$
   This removes the need to compute the normalizing constant.
3. **Sample a new $\theta$** using the slice sampling procedure.

By using slice sampling instead of Metropolis-Hastings, we avoid the need for proposal distributions and accept/reject steps, making it efficient in some cases.

## Gibbs Sampling with Slice Sampling
When using slice sampling within a **Gibbs sampler**, the steps are:
1. **Fix the other parameters** and sample the current parameter using slice sampling.
2. **Use the updated parameter** in the next Gibbs step.
3. **Repeat the cycle** until convergence.

This ensures that even if the full conditional distribution is not normalized, we can still sample from it efficiently.
